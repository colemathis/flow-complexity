```r
library(tidyverse)
```

```r
parameter.sweep.df <- read.csv("../data/2022_07_22_parameter_sweep_stats.csv", row.names=1)
```

```r
p <- ggplot(parameter.sweep.df %>% filter(N_reactors == 25)) + 
        geom_tile(aes(x= ave_forward_rate*mass, 
                      y = ave_outflow_rate,
                      fill = alpha,
                      width = 0.25, height = 0.25)) + 
        scale_y_log10() + 
        scale_x_log10() + 
        theme_minimal() 

p
```

```r
p <- ggplot(parameter.sweep.df %>% filter(ave_outflow_rate ==10.0 )) + 
        geom_point(aes(x= ave_forward_rate, 
                       y = median.AI,
                       color = as.factor(N_reactors), 
                       shape = graph_type), size = 2 ) + 
        scale_x_log10() + 
        theme_minimal() #+ facet_wrap(~graph_type)

p
```
